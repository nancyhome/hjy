```python
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm, rcParams
from torch.utils.tensorboard import SummaryWriter
from torchvision import transforms
```

```python
writer=SummaryWriter("logs")
```

```python
#ToTensor工具的使用
trans_tensor=transforms.ToTensor()
```

```python
img = Image.open("E:\Python\d2l-zh-1.0\hjy\lelush05.jpg")
writer.add_image("img",trans_tensor(img),1)

data1 = transforms.RandomResizedCrop(224)(img)
writer.add_image("img",trans_tensor(data1),2)
data2 = transforms.RandomResizedCrop(224)(img)
writer.add_image("img",trans_tensor(data2),3)
data3 = transforms.RandomResizedCrop(224)(img)
writer.add_image("img",trans_tensor(data3),4)
```

```python
#画图找不到字体的话
#显示中文标签
plt.rcParams['font.sans-serif']=['SimHei'] #显示中文标签
plt.rcParams['axes.unicode_minus']=False   #这两行需要手动设置
plt.subplot(2,2,1),plt.imshow(img),plt.title("原图")
plt.subplot(2,2,2),plt.imshow(data1),plt.title("转换后的图1")
plt.subplot(2,2,3),plt.imshow(data2),plt.title("转换后的图2")
plt.subplot(2,2,4),plt.imshow(data3),plt.title("转换后的图3")
plt.show()
```

```python
img_tensor=trans_tensor(img)
#Normalize
print(img_tensor[0][0][0])
trans_norm=transforms.Normalize([0.5,0.5,0.5],[0.5,0.5,0.5])
img_norm=trans_norm(img_tensor)
print(img_tensor[0][0][0])
writer.add_image("Normalize",img_norm,5)
```

```python
writer.close()
```
